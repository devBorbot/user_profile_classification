```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Load your CSV
df = pd.read_csv('./data/up_sample.csv')
df.head()
```

```python
# Aggregate by customer and category
pivot = df.pivot_table(
    index='ACCOUNTID', 
    columns='MATERIAL_PCAT_DESC', 
    values='PCAT_BILL_QTY',
    aggfunc='sum', 
    fill_value=0
).reset_index()
```

```python
# Merge back the target
targets = df[['ACCOUNTID', 'TARGET_1']].drop_duplicates()
dataset = pivot.merge(targets, on='ACCOUNTID', how='left')
```

```python
# Encode the TARGET_1 if needed
dataset['TARGET_1_CODED'] = LabelEncoder().fit_transform(dataset['TARGET_1'])
```
